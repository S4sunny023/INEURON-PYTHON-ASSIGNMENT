2) What is the relationship between generators and iterators?

An iterator is an object that implements the iterator protocol, which requires two methods: __iter__() and __next__(). The __iter__() method returns the iterator object itself, and the __next__() method returns the next value in the sequence.

Generators are a type of iterator that can be defined using a special syntax in Python, using the yield keyword. When a generator function is called, it returns a generator object that can be used to iterate over the sequence of values. Each time the yield keyword is encountered in the function, the current value is returned to the caller, and the state of the function is saved. 

all generators are iterators because they implement the iterator protocol, but not all iterators are generators. Generators are a special type of iterator that can be defined using the yield keyword and are more memory-efficient than other types of iterators.

3) What are the signs that a function is a generator function?

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1
```

In this example, countdown() is a generator function because it contains the yield keyword. When countdown() is called with an integer argument, it returns a generator object that can be used to iterate over the sequence of values generated by the function.

4) What is the purpose of a yield statement?

The purpose of the yield statement is to allow the generator function to produce a sequence of values without generating the entire sequence at once, as is the case with lists. Instead of creating a sequence of values and storing it in memory, generators compute the values on the fly as they are requested.
